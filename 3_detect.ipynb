```python
import SmartDetection as sd
from ultralytics import YOLO
import cv2

#Your best trained yolo model path and esp8266 url
model = YOLO("C:/Users/aurel/Documents/Yolo_Detection/runs/detect/train7/weights/best.pt")
#IP address of the esp8266
esp8266_url = "http://192.168.178.48/send"

#using SmartDetection class from SmartDetection.py for the math logic and esp8266 communication
detector = sd.SmartDetection(model, esp8266_url)

#Detect YOLO Model from webcam
cap = cv2.VideoCapture(0)
cv2.namedWindow("Smart Waste Detection", cv2.WINDOW_NORMAL)

while True:
    ret, frame = cap.read()
    if not ret:
        continue

    annotated = detector.process(frame)

    # Overlay detection progress
    if detector.current_label:
        text = f"{detector.current_label}: {detector.frame_count}/{detector.STABLE_FRAMES}"
        #To put text on the frame
        cv2.putText(
            annotated, text, (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2
        )

    cv2.imshow("Smart Waste Detection", annotated)
    # Break loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
